Here we don't Required " 'Day' , 'Month' , 'Year' " Based on These Independent Feature ['Temperature', 'RH', 'Ws', 'Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI'] I will try to Calculate What is 'FWI'  

Since it Linear regression Problem will conider all Independent Feature to calculate FWI evn we will consider ['Classes'='Independent Feature']

IF it was Classificatio Problem we would Considered Classes as Dependent Feature.

```python
## drop month day year 
df.drop(['day', 'month', 'year'], axis=1,inplace =True)
```

```python
df.head()
```

```python
df['Classes'].value_counts() ## Counts the no of occurance of the unique Value
```

```python
## Encoding Categorical Features

## Checks whether the text exists anywhere inside the string
## Works even if extra text is present
## Safer and more flexible
## “If the sentence has the words ‘not fire’ anywhere, treat it as 0.”

df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)  ## For 'not fire'=0 and 'Fire'=1
```

```python
df['Classes'].value_counts() ## Counts the no of occurance of the unique Value 
```
