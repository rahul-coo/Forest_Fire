# Data_Cleaning 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1) # here header=1 is used to skip the first row
```

```python
data.info()
```

Cleaning Start

```python
data.isnull().any(axis=1) # here it tell us that which row has null value and 
```

```python
data[data.isnull().any(axis=1)]
```

```python
# here we are assigning the new column with region values Dividing the dataset onto two region ie; region 0 and 1.

data.loc[:122,"Region"]=0 #index 0-122 assigned to region 0
data.loc[122:,"Region"]=1 #index 122-244 assigned to region 1 ; 122 is overridden.


df=data
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df=df.dropna().reset_index(drop=True) # here it reset the Index after dropping the null values 
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]] # it return 
```

```python
# Removing the 122 row and reseting the index
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip() # here it remove the extra spaces from the column names
```

```python
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O'] # here we are storing all the object type columns in a list
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```
